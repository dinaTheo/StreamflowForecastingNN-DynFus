```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mul_ensemble_model1 = pd.read_csv("durations-and-metrics-encdec-mul-dyn-fusion.csv")
mul_ensemble_model2 = pd.read_csv("durations-and-metrics-encdec-mul-dyn1-fusion.csv")

mul_ensemble_model1['duration'] = mul_ensemble_model1['duration']/3600
mul_ensemble_model2['duration'] = mul_ensemble_model2['duration']/3600

un_ensemble_model1 = pd.read_csv("durations-and-metrics-encdec-un-dyn-fusion.csv")
un_ensemble_model2 = pd.read_csv("durations-and-metrics-encdec-un-dyn1-fusion.csv")

un_ensemble_model1['duration'] = un_ensemble_model1['duration']/3600
un_ensemble_model2['duration'] = un_ensemble_model2['duration']/3600

mul_ensemble_nse = pd.concat([
    mul_ensemble_model1[['NSE']].assign(Model="Model 1", Approach="Multivariate"),
    mul_ensemble_model2[['NSE']].assign(Model="Model 2", Approach="Multivariate"),
])

un_ensemble_nse = pd.concat([
    un_ensemble_model1[['NSE']].assign(Model="Model 1", Approach="Univariate"),
    un_ensemble_model2[['NSE']].assign(Model="Model 2", Approach="Univariate"),
])

mul_ensemble_mae = pd.concat([
    mul_ensemble_model1[['MAE']].assign(Model="Model 1", Approach="Multivariate"),
    mul_ensemble_model2[['MAE']].assign(Model="Model 2", Approach="Multivariate"),
])

un_ensemble_mae = pd.concat([
    un_ensemble_model1[['MAE']].assign(Model="Model 1", Approach="Univariate"),
    un_ensemble_model2[['MAE']].assign(Model="Model 2", Approach="Univariate"),
])

mul_ensemble_maeh = pd.concat([
    mul_ensemble_model1[['MAE_HIGH']].assign(Model="Model 1", Approach="Multivariate"),
    mul_ensemble_model2[['MAE_HIGH']].assign(Model="Model 2", Approach="Multivariate"),
])

un_ensemble_maeh = pd.concat([
    un_ensemble_model1[['MAE_HIGH']].assign(Model="Model 1", Approach="Univariate"),
    un_ensemble_model2[['MAE_HIGH']].assign(Model="Model 2", Approach="Univariate"),
])

mul_ensemble_time = pd.concat([
    mul_ensemble_model1[['duration']].assign(Model="Model 1", Approach="Multivariate"),
    mul_ensemble_model2[['duration']].assign(Model="Model 2", Approach="Multivariate"),
])

un_ensemble_time = pd.concat([
    un_ensemble_model1[['duration']].assign(Model="Model 1", Approach="Univariate"),
    un_ensemble_model2[['duration']].assign(Model="Model 2", Approach="Univariate"),
])

mul_ensemble_pred_time = pd.concat([
    mul_ensemble_model1[['prediction-duration']].assign(Model="Model 1", Approach="Multivariate"),
    mul_ensemble_model2[['prediction-duration']].assign(Model="Model 2", Approach="Multivariate"),
])

un_ensemble_pred_time = pd.concat([
    un_ensemble_model1[['prediction-duration']].assign(Model="Model 1", Approach="Univariate"),
    un_ensemble_model2[['prediction-duration']].assign(Model="Model 2", Approach="Univariate"),
])

nse_data = pd.concat([mul_ensemble_nse, un_ensemble_nse])
mae_data = pd.concat([mul_ensemble_mae, un_ensemble_mae])
maeh_data = pd.concat([mul_ensemble_maeh, un_ensemble_maeh])

time_data = pd.concat([mul_ensemble_time,un_ensemble_time]).reset_index(drop=True)
time_pred_data = pd.concat([mul_ensemble_pred_time,un_ensemble_pred_time]).reset_index(drop=True)
```

```python
sns.set_theme(style="white", palette="muted")

fig, axes = plt.subplots(3, 1, figsize=(15, 23), sharey=False, dpi=500)

colors = {
    "Multivariate": "#6EAFF0",
    "Univariate": "dodgerblue"   
}

sns.boxplot(ax=axes[0], x="Model", y="MAE", hue="Approach", data=mae_data,
            palette=colors, linewidth=2, fliersize=6, showfliers=True)
axes[0].set_xlabel("", fontsize=30, fontweight="bold", labelpad=20)
axes[0].set_ylabel("MAE", fontsize=50, labelpad=35, fontweight="bold")
axes[0].tick_params(axis='x', labelsize=35)
axes[0].tick_params(axis='y', labelsize=35)
axes[0].legend_.remove()

sns.boxplot(ax=axes[1], x="Model", y="duration", hue="Approach", data=time_data,
            palette=colors, linewidth=2, fliersize=6, showfliers=True)
axes[1].set_xlabel("", fontsize=30, fontweight="bold", labelpad=20)
axes[1].set_ylabel("Training time (h)", fontsize=50, labelpad=7, fontweight="bold")
axes[1].tick_params(axis='x', labelsize=35)
axes[1].tick_params(axis='y', labelsize=35)
axes[1].legend_.remove()

sns.boxplot(ax=axes[2], x="Model", y="prediction-duration", hue="Approach", data=time_pred_data,
            palette=colors, linewidth=2, fliersize=6, showfliers=True)
axes[2].set_xlabel("", fontsize=30, fontweight="bold", labelpad=20)
axes[2].set_ylabel("Inference time (s)", fontsize=50, labelpad=35, fontweight="bold")
axes[2].tick_params(axis='x', labelsize=35)
axes[2].tick_params(axis='y', labelsize=35)
axes[2].legend_.remove()

fig.legend(labels=axes[0].get_legend_handles_labels()[1],
           handles=axes[0].get_legend_handles_labels()[0], 
           loc="lower center", ncol=2, fontsize=45, title_fontsize=26, bbox_to_anchor=(0.5, 0 - 0.1))

for ax in axes:
    for spine in ax.spines.values():
        spine.set_color("black")
        spine.set_linewidth(2)

plt.tight_layout(rect=[0, 0, 1, 0.95])
fig.supxlabel("Models", fontsize=45, x=0.55, y=-0.02, fontweight="bold")
plt.text(0.27, 3.35, "\nModel 1: LSTM-EncDec-Att-Fus\nModel 2: LSTM-EncDec-Op-Lev-Fus", fontsize=40, ha='left', transform=plt.gca().transAxes)
plt.show()
```
